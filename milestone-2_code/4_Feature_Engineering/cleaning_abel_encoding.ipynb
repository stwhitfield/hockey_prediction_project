```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Read in data and assign X and y
#data = pd.read_csv('../data/train_old.csv', index_col=0)
#data = pd.read_csv('../data/test_regular_old.csv', index_col=0)
data = pd.read_csv('../data/test_playoff_old.csv', index_col=0)
data
```

```python
data.isnull().sum(axis = 0)
```

```python
data.dropna(inplace=True)
data
```

```python
print(data['shotType'].unique())
print(data['shotType'].value_counts())
print(data['lastEventType'].unique())
print(data['lastEventType'].value_counts())
```

```python
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
data['shotType'] = labelencoder.fit_transform(data['shotType'])
data['lastEventType'] = labelencoder.fit_transform(data['lastEventType'])
data
```

```python
print(data['shotType'].unique())
print(data['shotType'].value_counts())
print(data['lastEventType'].unique())
print(data['lastEventType'].value_counts())
```

```python
data["rebound"] = data["rebound"].astype(int)
data
```
