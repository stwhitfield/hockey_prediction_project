```python
import pandas as pd

df=pd.read_csv("tidy.csv")

df.head()

data=df[df["eventTeam"]=="Colorado Avalanche"]
data.head()
```

```python
import random

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from PIL import Image, ImageDraw, ImageFont

# Fix Coordinates
data.loc[data['coordinateX'] < 0, ['coordinateX', 'coordinateY']] *= -1


MY_DPI = 96
IMG_WIDTH = 500
IMG_HEIGHT = 300
fig = plt.figure(figsize=(IMG_WIDTH / MY_DPI, IMG_HEIGHT / MY_DPI), dpi=MY_DPI)
ax = fig.add_subplot(111, frameon=False, xticks=[], yticks=[])

ax_extent = [-100, 100, -42.5, 42.5]
img = Image.open('Half_ice_hockey_rink-right.png')
plt.imshow(img, extent=ax_extent)

# Draw the seaborn portion of the graph
sns.set_style("white")
sns.kdeplot(data.coordinateX, data.coordinateY, cmap='bwr', shade=True, shade_lowest=False, alpha=0.6)

# Hide all axes & bounding boxes
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.set_frame_on(False)
ax.axis('off')

plt.show()
```
